#### **Valid Sales, Pending Orders, Cancelled Orders, Returned Orders, Revenue According To Order Status**

```python
valid_sales = df[df['Status'].isin(['Shipped', 'Shipped - Delivered to Buyer', 'Shipped - Picked Up', 'Shipped - Out for Delivery', 'Shipping'])].copy()
valid_sales = valid_sales[(valid_sales['Qty'] > 0) & (valid_sales['Amount'] > 0)]
pending_sales = df[df['Status'].isin(['Pending', 'Pending - Waiting for Pick Up'])].copy()
cancelled_sales = df[df['Status'] == 'Cancelled'].copy()
returned_sales = df[df['Status'].isin(['Shipped - Lost in Transit', 'Shipped - Damaged', 'Shipped - Rejected by Buyer', 'Shipped - Returning to Seller', 'Shipped - Returned to Seller'])].copy()

revenue = valid_sales[valid_sales['Amount'] > 0]
total_revenue = revenue['Amount'].sum()
total_pending_amount = pending_sales['Amount'].sum()
cancelled_amount = cancelled_sales['Amount'].sum()
returned_amount = returned_sales['Amount'].sum()
```

**During data validation, 101 records were identified with Status = 'Shipped' but Qty = 0 and Amount = 0. Since these records do not represent completed revenue-generating transactions, they were excluded from the sales analysis.**

```python
print('Total Revenue: ${:.2f}'.format(total_revenue))
print('Total Pending Amount: ${:.2f}'.format(total_pending_amount))
print('Total Cancelled Amount: ${:.2f}'.format(cancelled_amount))
print('Total Products Returned Amount: ${:.2f}'.format(returned_amount))
```

#### **Check Revenue Outliers**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(revenue, x=revenue['Amount'], color='#99004c')
plt.xlabel('Revenue')
plt.show()
```

```python
q1 = revenue['Amount'].quantile(0.25)
q3 = revenue['Amount'].quantile(0.75)
IQR = q3 - q1
lower = q1 - 1.5 * IQR
upper = q3 + 1.5 * IQR
print('Lower Fence: ',lower)
print('Upper Fence', upper)
outliers = revenue[(revenue['Amount'] < lower) | (revenue['Amount'] > upper)]
print('Outliers Shape: ',outliers.shape)
```

```python
outlier_percent = len(outliers) / len(revenue) * 100
print('Outlier Is {:.2f}% Of Total Revenue'.format(outlier_percent))
```

```python
visu_data = {
    'Metric': ['Revenue', 'Pending Orders Amount', 'Cancelled Orders Amount', 'Returned Orders Amount'], 
    'Data': [total_revenue, total_pending_amount, cancelled_amount, returned_amount]
}
datas = pd.DataFrame(visu_data)
sns.set_theme(style='whitegrid')
plt.figure(figsize=(12, 6))
x = sns.barplot(
    data=datas, 
    y='Metric', 
    x='Data', 
    hue='Metric', 
    palette='Paired',
    edgecolor='black',
    linewidth=0.1,
    legend=False
)
for container in x.containers:
    x.bar_label(
        container, 
        fmt='$%1.2f',      
        padding=8,         
        weight='bold', 
        color='#2d3748', 
        fontsize=12
    )
plt.title("Financial Overview Summary", fontsize=20, weight="bold", pad=5)
plt.xlabel('Amount ($)', fontsize=15, weight='bold')
plt.ylabel('')
plt.xlim(0, max(datas['Data']) * 1.15) 
plt.tight_layout()
plt.show()
```

#### **Growth Rate [March - April, April - May, May - June]**

```python
plt.figure(figsize=(10,6))
gd = {'growth_mo':['March To April', 'April To May', 'May To June'], 'per': [growth_rate_march_to_april, growth_rate_april_to_may,
                                                                            growth_rate_may_to_june]}
growth = pd.DataFrame(gd)
maxi = growth['per'].max()
sns.lineplot(growth, x = 'growth_mo', y = 'per', linewidth = 2.5, color='#ff0075', mec='#ff0075',markerfacecolor='#ff0075', marker='o')
plt.axhline(y=maxi, color='#00cc46', linestyle='--', linewidth=1.5)
plt.text(x=0.2, y=maxi - (maxi * 0.04), s=f'Peak Growth: {maxi:.2f}%', color='black', weight='bold', fontsize=11)
plt.fill_between(growth['growth_mo'], growth['per'], alpha=0.2, color='#ff0075')
plt.xlabel('Growth Months', fontsize=15, weight='bold')
plt.ylabel('Growth Percentage', fontsize=15, weight='bold')
plt.title('Growth Percentages Over Months', fontsize=20, weight='bold')
plt.tight_layout()
plt.show()
```

#### **Pareto Analysis(80/20) Rule**

```python
pareto = (revenue.groupby('Category')['Amount'].sum().sort_values(ascending=False)).reset_index()
pareto['Cumulative Revenue'] = pareto['Amount'].cumsum()
pareto['Cumulative Percentage'] = (pareto['Cumulative Revenue'] / pareto['Amount'].sum()) * 100
pareto.style.background_gradient(cmap='summer_r').format({'Cumulative Revenue': '${:.2f}'})
```

```python
fig, ax1 = plt.subplots(figsize=(12,6))
sns.barplot(data=pareto, x='Category', y='Amount', hue='Category', palette='Set2', ax=ax1)
for c in ax1.containers:
    ax1.bar_label(c, weight='bold', padding=2, fmt='%1.2f')
ax1.set_xlabel('Category', fontsize=14, weight='bold')
ax1.set_ylabel('Revenue', fontsize=14, weight='bold')
ax1.set_title('Pareto Chart of Revenue by Category', fontsize=18, weight='bold')
ax2 = ax1.twinx()
ax2.plot(pareto['Category'], pareto['Cumulative Percentage'], color='red', marker='o', linewidth=2, mfc='black', mec='black')
ax2.axhline(80, color='gray', linestyle='--', linewidth=1.5)
ax2.set_ylabel('Cumulative Percentage (%)', fontsize=14, weight='bold')
ax2.set_ylim(0, 110)
plt.tight_layout()
plt.show()
```

#### **Insights & Recommendations**

There were a considerable number of records where the order status was marked as "Shipped" but the order amount was zero. Since these records could distort revenue-related analyses, they were removed during the data cleaning process. I made every effort to clean the dataset while preserving as much useful information as possible.

Overall, this was an excellent dataset to work with due to its large size and variety of features. I enjoyed exploring it and performing the analysis.

The dataset contains sales data from March to June. March contains significantly fewer records because the dataset begins near the end of the month. Although I considered excluding March from the analysis, I decided to keep it for completeness.

From the monthly analysis, I observed that the number of orders remained relatively stable from April to June, while total revenue showed a slight downward trend. This suggests that although customers continued placing orders at a similar rate, the average revenue generated per order may have decreased.

Among all product categories, Set, Kurta, and Western Dress generated the highest revenue by a significant margin. These categories consistently outperformed the others and appear to be the primary contributors to overall sales. If this trend continues over a longer period, Amazon may benefit from giving additional attention to these high-performing categories through inventory planning, promotions, or product expansion.

The analysis also shows that B2B orders account for less than 1% of all orders, indicating that the platform is heavily dominated by B2C customers. This may represent an opportunity for Amazon to expand its B2B presence, provided there is sufficient market demand. However, further analysis would be required to determine whether increasing B2B sales would significantly improve overall revenue.

Geographically, Bengaluru recorded the highest number of orders among cities, while Maharashtra led among states. These regions represent strong markets, while other locations may offer opportunities for future growth.

Overall, the order volume remained relatively consistent throughout the analyzed months, whereas revenue experienced a gradual decline. This pattern suggests that monitoring customer demand, pricing strategies, and inventory management could help maintain or improve revenue performance in future periods.

This project provided valuable insights into Amazon's sales performance and demonstrated how exploratory data analysis can uncover meaningful business patterns from large real-world datasets.
